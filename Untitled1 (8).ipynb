```python
import torch
import cv2
import random
import glob
import pandas as pd
import matplotlib.pyplot as plt
from IPython.display import display, Image
```

```python
#Load the pre-trained YOLOv5 model
yolo_model = torch.hub.load("ultralytics/yolov5", "yolov5s")
```

```python
# Define object detection function
def detect_objects(image_path, details=False):
  image = cv2.imread(image_path)# Load image
  results = yolo_model(image, size=200)# Run inference
  results.print() # Print results summary
  results.show()# Display image
  if details:
    display(results.pandas().xyxy[0])
  return results.pandas().xyxy[0]['name'].tolist()
```

```python
#Run detection on 5 random images
print("\n--- Running object detection on 5 sample images ---")
sample_image_path = random.sample(image_paths, 5)
detection_results = []

for i in range(5):
    labels = detect_objects(sample_image_path[i], details=True)
    detection_results.append({
        "image": os.path.basename(sample_image_path[i]),
        "labels": labels
    })

print("\n--- Object Detection Results ---")
for result in detection_results:
    print(f"Image: {result['image']}")
    print(f"Detected Objects: {', '.join(result['labels']) if result['labels'] else 'None'}\n")
```
